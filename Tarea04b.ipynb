```python
import pandas as pd

# 1. Cargar el archivo CSV
try:
    df = pd.read_csv('archivo.csv')
    print("Archivo cargado correctamente. Columnas disponibles:")
    print(df.columns.tolist())  # Muestra todas las columnas disponibles
    
    # 2. Verificar y seleccionar la columna numérica correcta
    nombre_columna = 'ingresos'  # REEMPLAZA 'ingresos' con el nombre de tu columna numérica real
    
    if nombre_columna in df.columns:
        # 3. Convertir a numérico por si hay datos no numéricos
        df[nombre_columna] = pd.to_numeric(df[nombre_columna], errors='coerce')
        
        # 4. Calcular máximo y mínimo excluyendo valores NaN
        maximo = df[nombre_columna].max(skipna=True)
        minimo = df[nombre_columna].min(skipna=True)
        
        # 5. Mostrar resultados
        print(f"\nEstadísticas de '{nombre_columna}':")
        print(f"Máximo: {maximo}")
        print(f"Mínimo: {minimo}")
        print(f"Cantidad de valores no numéricos: {df[nombre_columna].isna().sum()}")
    else:
        print(f"\nError: La columna '{nombre_columna}' no existe.")
        
except FileNotFoundError:
    print("Error: No se encontró el archivo 'archive.csv'")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
import pandas as pd

try:
    # 1. Cargar el archivo CSV
    df = pd.read_csv('archivo.csv')
    
    # 2. Verificar columnas disponibles
    print("Columnas disponibles en el DataFrame:")
    print(df.columns.tolist())
    
    # 3. Especificar el nombre correcto de la columna (REEMPLAZAR ESTO)
    columna_numerica = 'ingresos'  # Cambia 'ingresos' por el nombre real de tu columna
    
    if columna_numerica in df.columns:
        # 4. Convertir a numérico y eliminar NaNs
        serie_numerica = pd.to_numeric(df[columna_numerica], errors='coerce').dropna()
        
        if not serie_numerica.empty:
            # 5. Calcular cuartiles y rango intercuartílico
            Q1 = serie_numerica.quantile(0.25)
            Q3 = serie_numerica.quantile(0.75)
            IQR = Q3 - Q1
            
            # 6. Calcular límites para outliers
            limite_inferior = Q1 - 1.5 * IQR
            limite_superior = Q3 + 1.5 * IQR
            
            # 7. Filtrar outliers
            mascara_outliers = (df[columna_numerica] < limite_inferior) | (df[columna_numerica] > limite_superior)
            outliers = df[mascara_outliers]
            
            # 8. Mostrar resultados
            print(f"\nAnálisis de outliers para '{columna_numerica}':")
            print(f"Primer cuartil (Q1): {Q1:.2f}")
            print(f"Tercer cuartil (Q3): {Q3:.2f}")
            print(f"Rango intercuartílico (IQR): {IQR:.2f}")
            print(f"Límite inferior: {limite_inferior:.2f}")
            print(f"Límite superior: {limite_superior:.2f}")
            print("\nValores atípicos encontrados:")
            print(outliers.to_string(index=False))
        else:
            print(f"\nLa columna '{columna_numerica}' no contiene valores numéricos válidos")
    else:
        print(f"\nError: La columna '{columna_numerica}' no existe en el DataFrame")

except FileNotFoundError:
    print("Error: No se encontró el archivo 'archivo.csv'")
except Exception as e:
    print(f"Ocurrió un error inesperado: {str(e)}")
```
